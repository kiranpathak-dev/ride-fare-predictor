#### Investigate the variables
Sort and interpret the data table for two variables:`trip_distance` and `total_amount`.

```python
# Sort the data by trip distance from maximum to minimum value

trip_distance = df.sort_values(by=['trip_distance'],ascending=False)
trip_distance.head(10)

# Sort the data by trip distance from min to max 
# trip_distance = df.sort_values(by=['trip_distance'],ascending=True)
# trip_distance.head(10)
```

The values align with our earlier data discovery, where we noticed that the longest rides are approximately 33 miles.

```python
# Sort the data by total amount from maximum to minimum value and print the top 20 values

total_amount_sorted = df.sort_values(['total_amount'],ascending=False)['total_amount']
total_amount_sorted.head(20)
```

```python
# Sort the data by total amount and print the bottom 20 values
total_amount_sorted.tail(20)
```

```python
# How many of each payment type are represented in the data
df['payment_type'].value_counts()
```

```python
# average tip for trips paid for with credit card
avg_credit_card_tip = df[df['payment_type'] == 1]['tip_amount'].mean()
print(f"Average credit card tip: {avg_credit_card_tip}")

# average tip for trips paid for with cash
avg_cash_tip = df[df['payment_type'] == 2]['tip_amount'].mean()
print(f"Average cash tip: {avg_cash_tip}")
```

```python
df['VendorID'].value_counts()
```

```python
# average total amount for each vendor
df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

```python
# Filter the data for credit card payments only
credit_card_passenger = df[df['payment_type'] == 1]

# Filter the data for credit card users by passenger count 
credit_card_passenger['passenger_count'].value_counts()
```

```python
# average tip amount for each passenger count (credit card payments only)
credit_card_passenger.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
```

After looking at the dataset, the two variables that are most likely to help build a predictive model for taxi ride fares are **total_amount** and **trip_distance** because those variables show a picture of a taxi cab ride.
